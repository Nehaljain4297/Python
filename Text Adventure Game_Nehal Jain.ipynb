__Story:__

Tumbbad is a haunting Indian horror film that explores themes of greed, obsession, and the supernatural. The story follows a young man who returns to his ancestral village of Tumbbad, a place shrouded in mystery and fear. Driven by his desire for wealth, he seeks out a hidden treasure said to be guarded by a terrifying creature 'Hastar'. As he delves deeper into the mystery, he becomes consumed by his greed and faces the consequences of his obsession.

__Game Introduction:__

This text-based adventure game will test your courage and cunning as you navigate the treacherous paths of an ancient, cursed kingdom in search of the treasure of Tumbbad, 'Gold Coins.' You find yourself standing at the entrance to a hidden palace. The palace is filled with ancient symbols and strange artifacts.

When you enter the palace, you encounter a monstrous creature, 'Hastar,' guarding the treasure chamber. Your task is to defeat Hastar and enter the treasure chamber to collect the gold coins. Once you have successfully defeated Hastar and collected the gold coins, you reach the exit gate, where a riddle is inscribed on it. You have three attempts to give the right answer to exit the palace or be trapped there forever.

__Objective of the game:__

The objective of the game is to guide the player through a perilous journey into the depths of Tumbbad ultimately escaping alive. The game is designed to test the player's courage, cunning, and problem-solving abilities as they navigate through dangerous challenges and make critical decisions that will determine their fate.


```python
# Prompt a user to play the game.
def stage_1():
    print("\nHold your breath. Let's start the text adventure game!")
    print("\nYou arrive at the ruins of Tumbbad. A mysterious door stands before you.")
    print()
    choice = input("Do you *enter* the palace or *turn back*?\n").lower()
    print()
    if choice == "enter":
        print("You step into the darkness...")
        print()
        stage_2()
    elif choice == "turn back":
        print("Fear grips your heart. You decide to walk back.")
        print()
        fail()
    else:
        print("Invalid choice. Please enter 'enter' or 'turn back'.")
        print()
        stage_1()
        

# In stage 2, the user attempts to enter the palace
def stage_2():

    choice = input("The palace door is filled with ancient symbols and strange artifacts. Do you *examine* the door or *search* the room?\n").lower()
    if choice == "examine":
        print("The door appears to be sealed shut. You need to find a way to open the door!")
        print()
        stage_2()
    elif choice == "search":
        items = ["map", "rusted key", "broken piece"]  # List of items found in the palace
        for item in items:
            if item == "map":
                print()
                print(f"You found a {item}! It will help you later!")
                print()
        print("You also found a hidden lever behind a statue. Pull it!!!")
        print()
        print("You pull the lever, and the heavy door creaks open, revealing a hidden chamber.")
        print()
        stage_3()
    else:
        print("Invalid choice. Please enter 'examine' or 'search'.")
        print()
        stage_2()
        
    
# Prompt the user to fight with Hastar. The outcome of the fight is random.         
def stage_3():
    print("You encounter a monstrous creature 'Hastar' guarding the treasure chamber.")
    print()
    choice = input("Do you want to fight with Hastar? (Y/N) \n")
    print()
    if choice.lower() == "y":
        print("You engage in a fierce battle with Hastar. The outcome is uncertain 'random'.")
        print()
        # Simulating a random battle outcome
        import random
        if random.randint(1, 3) == 1:
            print("You defeated Hastar and collected all the gold coins.")
            print()
            stage_4()
        else:
            print("Unfortunately, You couldn't fight well with Hastar.")
            fail()
    else:
        print("You got truly scared by Hastar. Show some courage!")
        fail()
        
        
# After the Hastar is defeated, the player solves a riddle inscribed on the exit gate. Player has 3 attempts to solve it.       
def stage_4():
    print("Now, you stand before the exit gate. A riddle is inscribed on it. \nYou have 3 attempts to give the right answer to exit the palace.")
    print()
    riddle_answer = input("'I have cities, but no houses. \nI have mountains, but no trees. \nI have water, but no fish.\nWho am I?'\n").lower()
    attempt=3
    while attempt>0:
        if riddle_answer == "map":
            print()
            print("Congratulations. You are a riddle master and the gold is all yours!")
            print()
            win()
            break 
            
        else:
            attempt-=1
            if attempt ==0:
                print()
                print("You couldn't solve the riddle. You are trapped in the palace forever")
                print()
                fail()
            else:
                print(f"You have {attempt} attempts left to guess the right answer.")
                riddle_answer = input("'I have cities, but no houses. \nI have mountains, but no trees. \nI have water, but no fish.\nWho am I?'\n").lower()
    
# Win function is executed when user defeats Hastar and exit the palace  
def win():
    print("You have successfully acquired the treasure of Tumbbad.")

# Fail function is executed when user loses against Hastar or fails to solve the riddle.
def fail():
    try:
        print("You have failed your quest. The darkness of Tumbbad has consumed you.")
        choice = input("Do you want to play again? (Y/N)\n").lower()
        if choice.lower() == "y":
            stage_1()
        elif choice.lower() == "n":
            print("You got truly scared by Hastar. Show some courage!")
        else:
            raise Exception("Error")
    except Exception :
        print("Invalid entry. Please enter (Y/N).")
        fail()
stage_1()
```
